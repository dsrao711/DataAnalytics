## Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

## Functional Plot

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1**2
plt.plot(x_1 , y_1)
plt.xlabel('Days')
plt.ylabel('Days squared')
plt.title('Days Squared Chart')
```

## Subplot

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

## Figure Object

```python
fig_1 = plt.figure(figsize = (7,4) , dpi = 100)
axes_1 = fig_1.add_axes([0.1 , 0.1 , 0.9 , 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Plot')
# Adding two plots in 1 graph
axes_1.plot(x_1 , y_1 , label = 'x/x**2')
axes_1.plot(y_1 , x_1 , label = 'x**/x')
axes_1.legend(loc = 0)

# Adding a small plot inside the figure
axes_2 = fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Small Days Squared Plot')
axes_2.plot(x_1 , y_1 , 'g')

#Adding text 
axes_2.text(0,40,'Message')
```

## Subplots

```python
fig_2, axes_2 = plt.subplots(figsize = (10,4) , nrows = 1 ,ncols = 3 )
#figsize(width , height)
#Spaces between plots
plt.tight_layout()
#Uses Indexing to acess plots
axes_2[1].set_title('Squares')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x_1,y_1)
```

## Appearance

```python
fig_3 = plt.figure(figsize = (6,4))
axes_3 = fig_3.add_axes([0,0,1,1])
axes_3.plot(x_1,y_1,color='#3B84A7', alpha=.75, lw=2, ls='-.',
           marker='o', markersize=6, markerfacecolor='y',
           markeredgecolor='c', markeredgewidth=1)
axes_3.set_title('Appearance')
axes_3.set_xlim([0,3])
axes_3.set_ylim([0,10])
axes_3.grid(True , color = '#4B8078' , dashes = (5,2,1,2))
axes_3.set_facecolor('#FEE9FF')
```
