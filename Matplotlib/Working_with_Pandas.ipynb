## With Pandas df

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv('icecreamsales.csv')
df = df.sort_values(by = 'Temperature')
#Converting pandas to Numpy array
np_array = df.values

x_1 = np_array[:,0]
y_1 = np_array[:,1]

fig_1 = plt.figure(figsize = (6,4))

axes_1 = fig_1.add_axes([0,0,1,1])
axes_1.set_title('Ice Cream Sales vs. Temperature')
axes_1.set_xlabel('Temperature')
axes_1.set_ylabel('Ice Cream Sales')


## Adding annotation

axes_1.annotate('Good month' , xy = (78,520) , xytext = (60,500) ,arrowprops=dict(facecolor='black', shrink=0.05))

axes_1.plot(x_1,y_1)
plt.bar(x_1,y_1)
```
